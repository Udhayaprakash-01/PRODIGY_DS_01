```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path = "C:\\Users\\Udhaya\\Documents\\API_SP.POP.TOTL_DS2_en_csv_v2_3401680.csv"
df = pd.read_csv(file_path, skiprows=4, sep=',')
```

```python
df.head()
```

```python
year = '2020'
df_year = df[['Country Name', year]].dropna()
```

```python
df_year_sorted = df_year.sort_values(by=year, ascending=False).head(10)
```

```python
plt.figure(figsize=(10, 6))
plt.bar(df_year_sorted['Country Name'], df_year_sorted[year], color='skyblue')
plt.xticks(rotation=45, ha='right')
plt.xlabel('Country')
plt.ylabel('Population')
plt.title(f'Top 10 Most Populated Countries in {year}')
plt.tight_layout()
plt.show()
```
